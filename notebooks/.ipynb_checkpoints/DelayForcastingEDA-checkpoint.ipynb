```python
import numpy as np
import pandas as pd
import os
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from pandas.plotting import scatter_matrix

datafolder = "data\\raw"
airport = "JFK"
month = "Sept"
year = "2019"
dates = "1-28"
```

```python
fileName = os.path.join(datafolder,   airport+"_"+month+"_"+dates+"_"\
                        +year+"_v2.csv")
df = pd.read_csv(fileName)
df["date"]=pd.to_datetime(df["date"])
df.info()
```

```python
df_by_weekday_mean = df.groupby(df['date'].dt.weekday_name).mean()
df_by_weekday_std = df.groupby(df['date'].dt.weekday_name).std()/np.sqrt(4)

new_index = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
df_by_weekday_mean = df_by_weekday_mean.reindex(new_index)
df_by_weekday_std = df_by_weekday_std.reindex(new_index)
df_by_weekday_mean.head()
```

```python
df_by_hour_mean = df.groupby(df['Hour']).mean()
df_by_hour_std = df.groupby(df['Hour']).std()/np.sqrt(28)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.errorbar(df_by_hour_mean.index, df_by_hour_mean["% DelayedGateDepartures"],df_by_hour_std["% DelayedGateDepartures"], ecolor="red")
plt.xlabel("Hour of day")
plt.ylabel("% delayed departures")

ax2 = ax1.twinx()
ax2.errorbar(df_by_hour_mean.index, df_by_hour_mean["AverageMinutesOf DelayPer DelayedGateDeparture"],df_by_hour_std["AverageMinutesOf DelayPer DelayedGateDeparture"], color="green",ecolor="magenta")
#ax2.yaxis.tick_right()
#ax2.yaxis.set_label_position("right")
plt.ylabel("Average delay (minutes)")
```

```python
plt.errorbar(df_by_weekday_mean.index, df_by_weekday_mean["% DelayedGateDepartures"],df_by_weekday_std["% DelayedGateDepartures"], ecolor="red")
```

```python
df["weekday"]=df["date"].dt.weekday_name
df["weekday"]=df["weekday"].astype("category")
df_for_regression = df[["Hour","DeparturesFor MetricComputation","% DelayedGateDepartures",\
                       "AverageMinutesOf DelayPer DelayedGateDeparture","ArrivalsFor MetricComputation","% DelayedGateArrivals"\
                      ,"AverageMinutesOf DelayPer DelayedGateArrival", "weekday"]]
df_for_regression.rename(columns={"DeparturesFor MetricComputatio": "DeparturesForMetricComputatio",\
                   "% DelayedGateDepartures": "%DelayedGateDepartures",\
                  "AverageMinutesOf DelayPer DelayedGateDeparture":"AverageMinutesOfDelayPerDelayedGateDeparture",\
                  "ArrivalsFor MetricComputation":"ArrivalsForMetricComputation",\
                  "% DelayedGateArrivals":"%DelayedGateArrivals",\
                  "AverageMinutesOf DelayPer DelayedGateArrival":"AverageMinutesOfDelayPerDelayedGateArrival"}, inplace = True)
df_for_regression.info()
```

```python
PercentDelayinFuture5Hr = np.empty((len(df_for_regression),1))
PercentDelayinFuture5Hr[:] = np.nan 
PercentDelayinFuture10Hr = np.empty((len(df_for_regression),1))
PercentDelayinFuture10Hr[:] = np.nan 
MinuteDelayinFuture5Hr = np.empty((len(df_for_regression),1))
MinuteDelayinFuture5Hr[:] = np.nan 
MinuteDelayinFuture10Hr = np.empty((len(df_for_regression),1))
MinuteDelayinFuture10Hr[:] = np.nan 
for itr in df_for_regression.index:
    if itr<len(df_for_regression)-10:
        PercentDelayinFuture10Hr[itr]=df_for_regression["%DelayedGateDepartures"][itr+10]
        PercentDelayinFuture5Hr[itr]=df_for_regression["%DelayedGateDepartures"][itr+5]
        MinuteDelayinFuture10Hr[itr]=df_for_regression["AverageMinutesOfDelayPerDelayedGateDeparture"][itr+10]
        MinuteDelayinFuture5Hr[itr]=df_for_regression["AverageMinutesOfDelayPerDelayedGateDeparture"][itr+5]
    elif itr<len(df_for_regression)-5:
        PercentDelayinFuture5Hr[itr]=df_for_regression["%DelayedGateDepartures"][itr+5]
        MinuteDelayinFuture5Hr[itr]=df_for_regression["AverageMinutesOfDelayPerDelayedGateDeparture"][itr+5]
    else:
        pass
df_for_regression["PercentDelayinFuture5Hr"] = PercentDelayinFuture5Hr
df_for_regression["PercentDelayinFuture10Hr"] = PercentDelayinFuture10Hr
df_for_regression["MinuteDelayinFuture5Hr"] = MinuteDelayinFuture5Hr
df_for_regression["MinuteDelayinFuture10Hr"] = MinuteDelayinFuture10Hr
```

```python
df_for_regression.dropna(inplace=True)
df_for_regression.info()
```

```python
train_set, test_set = train_test_split(df_for_regression,test_size=0.1)
train_PercentDelayinFuture5Hr = train_set["PercentDelayinFuture5Hr"]
train_PercentDelayinFuture10Hr = train_set["PercentDelayinFuture10Hr"]
train_MinuteDelayinFuture5Hr = train_set["MinuteDelayinFuture5Hr"]
train_MinuteDelayinFuture10Hr = train_set["MinuteDelayinFuture10Hr"]

test_PercentDelayinFuture5Hr = test_set["PercentDelayinFuture5Hr"]
test_PercentDelayinFuture10Hr = test_set["PercentDelayinFuture10Hr"]
test_MinuteDelayinFuture5Hr = test_set["MinuteDelayinFuture5Hr"]
test_MinuteDelayinFuture10Hr = test_set["MinuteDelayinFuture10Hr"]
```

```python
cat_attributes = ["weekday"]
num_attributes = ["Hour", "DeparturesFor MetricComputation","AverageMinutesOfDelayPerDelayedGateDeparture","%DelayedGateDepartures",\
              "ArrivalsForMetricComputation","AverageMinutesOfDelayPerDelayedGateArrival", "%DelayedGateArrivals"]

num_pipeline = Pipeline([
    ("std_scaler", StandardScaler()),
     ])
full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attributes),
    ("cat", OneHotEncoder(), cat_attributes),
    ])

train_set_prepared = full_pipeline.fit_transform(train_set)
test_set_prepared = full_pipeline.transform(test_set)
```

```python
train_set.shape
```

```python
lin_reg = LinearRegression()
tree_reg = DecisionTreeRegressor()
forest_reg = RandomForestRegressor()
```

```python
scores = cross_val_score(forest_reg, train_set_prepared, train_MinuteDelayinFuture5Hr,
                        scoring = "r2",cv=3)  #neg_mean_squared_error
#np.sqrt(-scores)
scores
```

```python
from sklearn import svm
from sklearn.svm import SVR

clf = SVR(C=20 , epsilon=0.5)
clf.fit(train_set_prepared, train_MinuteDelayinFuture5Hr)
prediction = clf.predict(test_set_prepared)

from sklearn.metrics import r2_score
r2_score(test_MinuteDelayinFuture5Hr, prediction)
```
